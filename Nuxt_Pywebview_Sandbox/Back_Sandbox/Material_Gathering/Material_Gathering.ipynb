## Tool For gathering materials(stock data)
help with production of new methods

```python
import pandas as pd

class GatherMaterials():
    def __init__(self, source, name, sDate, eDate):
        self.source = source
        self.sDate = sDate
        self.eDate = eDate
        self.name = name
        self.prepare_data()
    def prepare_data(self):
        
        self.data = pd.read_csv(f'{self.source}', sep='\t', names=['time', 'open', 'high', 'low', 'close', 'volume'], parse_dates=['time'], index_col = 'time')
        
        self.data.rename(columns = {'close' : f'{self.name}'}, inplace=True)
        
        self.data.index = self.data.index.tz_localize('UTC')

        # Seperating date logic
        if self.sDate.lower() == 'start':
            self.sDate = self.data.index[0]
        else:
            self.sDate += ' 00:00:00+00:00' # <--- Localizing dates to be UTC; 

        if self.eDate.lower() == 'end':
            self.eDate = self.data.index[-1]
        else:
            self.eDate += ' 23:00:00+00:00'

        # Localizing dates so dates don't have offset conflict

        self.data = self.data.loc[self.sDate : self.eDate]
        
        return self.data

    def make_csv(self, name):
        reduce = self.data.loc[:, self.name]
        reduce.to_csv(f'{name}.csv')
        print(f'Successfully made into csv! Filename: {name}.csv')
```

```python
backData = GatherMaterials('./AUDUSD30-2.csv', 'AUDUSD', '2021-01-01', '2025-02-14')
```
